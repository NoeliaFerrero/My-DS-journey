# Simplificando la lectura de archivos de texto de ancho fijo y comparación de dataframes con distintos números de filas

```python
#Importando las librerías a usar
import pandas as pd
import xlwings as xw
```

## Comparación de dos archivos con igual número de filas

```python
#Estableciendo el ancho de las columnas
widths = [8,12,6,6,6,8,8,8,7,7,8,8,6,8,8,7,9,7,8,8]
```

```python
#Leer el primer archivo de texto
texto1 = pd.read_fwf("reporte_1.txt", skiprows=66, widths=widths)[:-2]
texto1.head()
```

```python
#Leer el nombre de las columnas
texto1.columns
```

```python
# Establecer las columnas de interes
columnas1 = ['BUS#', 'X-- NAME --X', 'SYMM I', 'SYMM I.1']
texto1 = texto1[columnas1]
texto1.head()
```

```python
#Leer el segundo archivo de texto
texto2 = pd.read_fwf("reporte_2.txt", skiprows=66, widths=widths)[:-2]
```

```python
# Establecer las columnas de interes
columnas2 = ['SYMM I', 'SYMM I.1']
texto2 = texto2[columnas2]
texto2.head()
```

```python
#Copiar la información al archivo excel
archivof = 'Compara.xlsx'
wb = xw.Book(archivof)
sht = wb.sheets['Compara']
sht.range('a4').options(header=False, index = False).value = texto1
sht.range('e4').options(header=False, index = False).value = texto2
wb.save()
```

## Comparación de dos archivos con distinto número de filas

```python
#Lectura del archivo con número distinto de filas
texto3 = pd.read_fwf("reporte_3.txt", skiprows=66, widths=widths)[:-2]
```

```python
#Establecer las columnas de interes
columnas3 = ['BUS#', 'SYMM I', 'SYMM I.1']
texto3 = texto3[columnas3]
texto3.head()
```

```python
#Cambiar el "index" del dataframe
texto1 = texto1.set_index('BUS#')
texto3 = texto3.set_index('BUS#')
```

```python
#Unir ambos dataframes
resultado = pd.concat([texto1, texto3], axis=1, sort=False)
resultado.head()
```

```python
resultado
```

```python
#Copiar la información al archivo excel
archivof2 = 'Compara2.xlsx'
wb2 = xw.Book(archivof2)
sht2 = wb2.sheets['Compara']
sht2.range('a4').options(header=False, index = True).value = resultado
wb2.save()
```
