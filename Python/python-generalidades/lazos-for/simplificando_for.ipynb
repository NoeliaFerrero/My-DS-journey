 # Simplificando el uso de los lazos for

 ## Uso simple de for
 Imprimir cada uno de los elementos de una lista

```python
for i in range(5):
    print('i es : ',i)
```

 ## Operando con listas
 Obtener la suma de los cuadrados de los

 elementos de una lista

 ### Forma tradicional

```python
x = [1, 3, 5, 7, 9]
suma_cuadrados = 0

for i in range(len(x)):
    suma_cuadrados += x[i]**2
    print(suma_cuadrados)
```

```python
x[0]
```

```python
x[2]
```

```python
suma_cuadrados
```

 ### Un poco más pythonista

```python
x = [1, 3, 5, 7, 9]
suma_cuadrados = 0

for y in x:
    suma_cuadrados += y**2
    print(suma_cuadrados)
```

 ### La forma pythonista
 [***List comprehension***](https://docs.python.org/3/tutorial/datastructures.html)

 #### "Una forma más concisa y de fácil lectura"

```python
x = [1, 3, 5, 7, 9]
suma_cuadrados = sum([y**2 for y in x])
print(suma_cuadrados)
```

```python
[y**2 for y in x]
```

```python
sum([y**2 for y in x])
```

 ## Uso de for e if
 ### Forma tradicional

```python
x = [2, 4, 1, 8, 12, 3, 10, 15, 20, 5]
lista = []

for i in x:
    if i > 4:
        lista.append(i)
lista
```

 ### La forma pythonista

```python
x = [2, 4, 1, 8, 12, 3, 10, 15, 20, 5]
lista = [i for i in x if i > 4]
lista
```

 ## Uso de for, if y else
 ### Forma tradicional

```python
lista = []
x = [-2, -4, 1, 8, -12, 3, -10, 15, 20, 5]

for i in x:
    if i < 0:
        lista.append(i)
    else:
        lista.append(i * -1)
lista
```

 ### La forma pythonista

```python
x = [-2, -4, 1, 8, -12, 3, -10, 15, 20, 5]
lista = [i if i < 0 else i * -1 for i in x]
lista
```
