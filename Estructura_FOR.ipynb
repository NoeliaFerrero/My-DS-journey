NOTA: Python es de tipado dinámico, esto quiere decir que no se requiere la definición del tipo de datos en las variables, el lenguaje auto-asigna el tipo de datos según el valor declarado.

```python
frutas = ['kiwi','mango','cereza']  # definimos una variable del tipo lista y le asignamos 3 elementos
for fruta in frutas:                # definimos un ciclo for para recorrer la lista
  print(fruta)                      # imprimimos por pantalla cada elementos de la lista
```

#Importar Librerias

```python
import datetime
from pytz import timezone
from pytz import all_timezones
```

📚📦 El modulo (ó paquete) Pytz trae la base de datos "Olson tz" a Python🐍 
Al importar dicho modulo, accedemos a las funcionalidades de conversión de fecha y hora. Tambien nos permite cálculos de zona horaria en nuestras aplicaciones de Python y crear instancias de fecha y hora conscientes de la zona horaria. 💥💪

📝Hay algunos atributos en el módulo Pytz para ayudarnos a encontrar las cadenas de zona horaria admitidas. El atributo **all_timezones** contiene la lista de todas las zonas horarias disponibles

![what](https://media.giphy.com/media/SqmkZ5IdwzTP2/giphy.gif)

##Listado de zonas horarias disponibles

```python
print(len(all_timezones))   # imprimimos la cantidad de elementos que tiene la lista, usando el metodo len
for zona in all_timezones:  # recorremos la lista con un for (notar que a la variable zona se le autoasigna cada elem de la lista)
    print(zona)
```

##Buscar zona horaria específica

```python
for zona in all_timezones: # uso de ciclo for + if 
    if 'Sal' in zona:
        print(zona)
```
